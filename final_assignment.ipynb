### 1. Implementation of Sorting Algorithm

```python
A = [5,416,54,21,6135,15,741,81238]
# Sorting function 
def selection_sort(A):
    # Traversethe array (A) with variable (i) from 0 to length of array (A)
    for i in range(len(A)):
        # finding the index of minimum element
        minimum = i
        # Traverse the array (A) with variable (j) from i+1 to lenth of the array (A)
        for j in range(i+1, len(A)):
            # check the small element 
            if (A[minimum] > A[j]):
                minimum = j
        # basic swapping 
        temp = A[i]
        A[i] = A[minimum]
        A[minimum] = temp
    return A

# call the sort function
selection_sort(A)   
```

### Write a program that returns the file type from a file name. The type of the file is determined from the extension. Initially, a list of values of the form "extension,type"(e.g. xls,spreadsheet; png,image) will be input.


```python
def fileType(extensions, files):
    extension_types = extensions.split(';')
    extensionDict = {}

    for ex in extension_types:
        extension, file_type = ex.split(',')
        extensionDict[extension] = file_type

    filesDict = {}
    for file in files:
        if '.' in file:
            extn = file.split('.')
            extn = extn[-1]
            filesDict[file] = extensionDict.get(extn, "unknown")
        else:
            filesDict[file] = "unknown"

    return filesDict

fileType("xls,spreadsheet;xlsx,spreadsheet;jpg,image", ["abc.jpg", "xyz.xls", "text.csv", "123"])
```

### Q3. Column Sorting, yay!

```python
def sort_dict_list(dict, sort):
    def get(item):
        return item[sort]
    return sorted(dict, key=get)


# Testing the function
list = [
    dict(fruits= "apple", colors= "blue"),
    dict(fruits = "mango", colors =  "red")
]
sort_dict_list(list, "colors")
```

### Q4. The power of one line - Given a dictionary, switch position of key and values in the dict,

```python
def switch_key_value(d):
    # Inverts the keys and values of a dictionary
    return {v: k for k, v in d.items()}

print(switch_key_value({
    "key1": "val1",
    "key2": "val2",
}))
```

### Q5. Common, Not Common Given 2 lists in input. Write a program to return the elements, which are common to both lists(set intersection) and those which are not common(set symmetric difference) between the lists.

```python
def common_and_not_common(list1, list2):
    set1 = set(list1)
    set2 = set(list2)
    common = set1.intersection(set2)
    common = list(common)
    not_so_common = set1.symmetric_difference(set2)
    not_so_common = list(not_so_common)
    return common, not_so_common

mainstream = ["One Punch Man","Attack On Titan","One Piece","Sword Art Online","Bleach","Dragon Ball Z","One Piece"]
must_watch = ["Full Metal Alchemist","Code Geass","Death Note","Stein's Gate","The Devil is a Part Timer!","One Piece","Attack On Titan"]

common_and_not_common(mainstream, must_watch)
```

### Q7. Calculate the factorial of a number using lambda function.

```python
from functools import reduce

def calculate_factorial(z):
    num = range(1, z + 1)
    prod = reduce(lambda x, y: x * y, num)
    return prod

# Testing the function
print(calculate_factorial(5))  # Output: 120
```

### Q9.Write a func that takes 3 args:

```python
from datetime import datetime

def check_date_difference(from_date: str, to_date: str, difference: int) -> bool:
    # parse the input date strings to datetime objects
    from_date_obj = datetime.strptime(from_date, '%y-%m-%d')
    to_date_obj = datetime.strptime(to_date, '%y-%m-%d')

    # calculate the difference between the two dates
    date_difference = abs((to_date_obj - from_date_obj).days)

    # check if the difference is less than the input difference
    return date_difference < difference

```

### Q10. Write a func that takes 2 args:

```python
from datetime import datetime, timedelta

def date_n_days_before(dat, n):
    date = datetime.strptime(dat, '%y-%m-%d')
    delta = timedelta(days=n)
    new_date = date - delta
    new_date = "{:%y-%m-%d}".format(new_date)
    return new_date
```
